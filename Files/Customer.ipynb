```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges']
```

```python
df.shape
```

```python
X= df.drop(['Churn', 'customerID'], axis=1)
y= df['Churn']
```

```python
X_cat= X.select_dtypes(include=['object'])
X_num= X.select_dtypes(exclude=['object'])
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore', dtype='int')
le= LabelEncoder()
```

```python
ohe_features=X_cat[['MultipleLines', 'InternetService', 'Contract', 'PaymentMethod']]
ohe.fit(ohe_features)
ohe_features_names=ohe.get_feature_names_out(ohe_features.columns)
```

```python
label_feateres=X_cat.drop(['MultipleLines', 'InternetService',  'Contract', 'PaymentMethod'], axis=1)
```

```python
for i in label_feateres:
    X_cat[i]=le.fit_transform(label_feateres[i])
X_cat['StreamingMovies']=le.fit_transform(X_cat['StreamingMovies'])
```

```python
X_cat
```

```python
ohe_df=pd.DataFrame(ohe.transform(ohe_features), columns=ohe_features_names)
X_cat=X_cat.drop(['MultipleLines', 'InternetService',  'Contract', 'PaymentMethod'], axis=1)
X_cat=pd.concat([X_cat, ohe_df], axis=1)
X_cat
```

```python
X_final= pd.concat([X_num, X_cat], axis=1)
X_final['TotalCharges']=X_final['TotalCharges'].fillna(0)
X_final.isnull().sum()
```

```python
y=le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV, cross_validate
```

```python
models={
    'rf': RandomForestClassifier(),
    'gbc': GradientBoostingClassifier(),
    'hgbc': HistGradientBoostingClassifier()
    
}
params={
    'rf': {'n_estimators': [100, 200,250], 'max_depth': [None, 10, 20],'min_samples_leaf': [2, 5, 10]},
    'gbc': {'n_estimators': [100,200,250], 'learning_rate': [ 0.1,0.2,0.5,0.8],'min_samples_leaf': [2, 5, 10]},
    'hgbc': {'max_iter': [100, 200,250], 'max_depth': [None, 10],'min_samples_leaf': [2, 5, 10]}
}
```

```python
for model,value in models.items():
    classifier=RandomizedSearchCV(models[model],params[model], cv=5, n_iter=10)
    classifier.fit(X_train, y_train)
    print(f'Best parameters for {model}: {classifier.best_params_}')
    print(f'Best score for {model}: {classifier.best_score_}')

    
```

```python
def train(model,X_train,y_train,y_test):
    cross_validate(model, X_train, y_train, cv=5, scoring='accuracy')
    model.fit(X_train,y_train)
    score=model.score(X_test,y_test)
    print('Score of the model:', score)

```

```python
rf=RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_leaf=5)
train(rf, X_train, y_train, y_test)
```

```python
gbc=GradientBoostingClassifier(n_estimators=200, learning_rate=0.1, min_samples_leaf=2)
train(gbc, X_train, y_train, y_test)
```

```python
hgbc=HistGradientBoostingClassifier(max_iter=100, max_depth=5, min_samples_leaf=2)
train(hgbc, X_train, y_train, y_test)
```

```python
y_pred=rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm= confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(cm)
```

```python
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
